```python
#Initializing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
#Reading output of data scraping
input = pd.read_csv('C:/Users/sanyo/OneDrive - Indian School of Business/Term1/DC/Office_Hour/get_phones_data/get_phones_data/Tourist_Destinations_Of_The_World_v4.csv', sep=',')
input.head(10)
```

```python
input['Top_Restaurants'].fillna("Not Applicable",inplace=True)
```

```python
input['Hotel'].fillna("Not Applicable",inplace=True)
input['Hotel_Ratings'].fillna("Not Applicable",inplace=True)
input['Hotel_No_of_Reviews'].fillna("Not Applicable",inplace=True)
```

```python
input['Preferance_Family_Holiday_Destination'] = input['Preferance_Family_Holiday_Destination'].map(lambda x: str(x).replace('%',''))
input['Preferance_Solo_Trip_Destination'] = input['Preferance_Solo_Trip_Destination'].map(lambda x: str(x).replace('%',''))
input['Preferance_Friends_Holiday_Destination'] = input['Preferance_Friends_Holiday_Destination'].map(lambda x: str(x).replace('%',''))
input['Preferance_Spouse/Partner_Holiday_Destination'] = input['Preferance_Spouse/Partner_Holiday_Destination'].map(lambda x: str(x).replace('%',''))
```

```python
input['Preferance_Family_Holiday_Destination'] = pd.to_numeric(input['Preferance_Family_Holiday_Destination'])
input['Preferance_Solo_Trip_Destination'] = pd.to_numeric(input['Preferance_Solo_Trip_Destination'])
input['Preferance_Friends_Holiday_Destination'] = pd.to_numeric(input['Preferance_Friends_Holiday_Destination'])
input['Preferance_Spouse/Partner_Holiday_Destination'] = pd.to_numeric(input['Preferance_Spouse/Partner_Holiday_Destination'])
```

```python
temp1=input.groupby(['Destination_Name'])['Preferance_Family_Holiday_Destination'].min().reset_index()
temp2=input.groupby(['Destination_Name'])['Preferance_Solo_Trip_Destination'].min().reset_index()
temp3=input.groupby(['Destination_Name'])['Preferance_Friends_Holiday_Destination'].min().reset_index()
temp4=input.groupby(['Destination_Name'])['Preferance_Spouse/Partner_Holiday_Destination'].min().reset_index()
```

```python
#Which is the most popular destinatination for a Family Holiday?

temp1.plot(x='Destination_Name', y='Preferance_Family_Holiday_Destination', kind='bar',figsize=(15,10))
plot.xlabel("Tourist Destinations")
plot.ylabel("Percentage of people choosing the destination as most preferred for family holiday")
plot.grid(True)
```

Insight 1:
Tower Of London is the most popular destination for a Family Holiday out of all the 14 Tourist Destinations in scope
Forbidden City and Kazan Kremlin are the least popular destinations for a Family Holiday out of all the 14 Tourist Destinations in scope

```python
input['Visitors_per_year'] = input['Visitors_per_year'].apply(str)
input['Visitors_per_year'] = input['Visitors_per_year'].map(lambda x: str(x).replace('+',''))
input['Visitors_per_year'] = input['Visitors_per_year'].map(lambda x: str(x).replace(',',''))

input['Visitors_per_year'] = pd.to_numeric(input['Visitors_per_year'])
```

```python
#What is the count of number of visitors per year for each of the tourist destination?

figure = plot.figure(figsize=(15,10))
sns.barplot(data = input, y = 'Visitors_per_year', x = 'Destination_Name')
plot.show()
```

```python
input['Hotel_No_of_Reviews'] = input['Hotel_No_of_Reviews'].apply(str)
input['Hotel_No_of_Reviews'] = input['Hotel_No_of_Reviews'].map(lambda x: str(x).replace('reviews',''))
input['Hotel_No_of_Reviews'] = input['Hotel_No_of_Reviews'].map(lambda x: str(x).replace('Not Applicable',''))
input['Hotel_No_of_Reviews'] = input['Hotel_No_of_Reviews'].map(lambda x: str(x).replace(',',''))
input['Hotel_No_of_Reviews'] = pd.to_numeric(input['Hotel_No_of_Reviews'])
```

```python
input['Destination_Number_Of_Reviews'] = input['Destination_Number_Of_Reviews'].apply(str)
input['Destination_Number_Of_Reviews'] = input['Destination_Number_Of_Reviews'].map(lambda x: str(x).replace('reviews',''))
input['Destination_Number_Of_Reviews'] = input['Destination_Number_Of_Reviews'].map(lambda x: str(x).replace(',',''))
input['Destination_Number_Of_Reviews'] = pd.to_numeric(input['Destination_Number_Of_Reviews'])
```

```python
#Plotting the number of destination reviews with the destination name
figure = plot.figure(figsize=(18,8))
sns.barplot(data = input, y = 'Destination_Number_Of_Reviews', x = 'Destination_Name')
plot.show()
```

Insight 10:
We notice that Sagrada Familia has the highest number of reviews followed by Eiffel Tower. Also, we notice that Forbidden City has very less reviews, although earlier in the earlier answerset, we noticed that Forbidden city had the highest number of visitors. This goes on to show that not necessarily every visitor gives reviews for their destinations visited. Thus, having the highest number of visitors does not necessarily mean that the visitors would have reviewed them

```python
#Plotting the destination reviews with the destination name along with destination rating
figure = plot.figure(figsize=(18,8))
sns.stripplot(data = input, y = 'Destination_Number_Of_Reviews', x = 'Destination_Name', hue = 'Destination_rating')
plot.show()
```

Insight 11:
We notice that Kazan Kremlin has a 5 rating but has the least number of reviews given by visitors, whereas Sagrada Familia has the highest number of reviews but with 4.5 rating. Thus, we can say that having highest number of reviews doesnt necessarily mean that the visitors have liked the place.
